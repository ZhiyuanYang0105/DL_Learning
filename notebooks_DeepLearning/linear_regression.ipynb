# 线性回归

## 什么是线性回归

### 线性回归（Linear Regression）定义

线性回归是一种用于刻画输入变量与输出变量之间线性关系的统计学习方法。

给定输入向量 $x \in \mathbb{R}^p$（指p维实数空间），输出为 $y \in \mathbb{R}$，线性回归模型假设：

$$
y = \mathbf{w}^T \mathbf{x} + b + \epsilon
$$

其中：

- $\mathbf{w} \in \mathbb{R}^p$：权重向量  
- $b \in \mathbb{R}$：偏置（截距）  
- $\epsilon$：随机误差项（通常假设为均值为 0 的噪声）

对于数据集：

$$
\{(x_i, y_i)\}_{i=1}^n（表示有n个数据）
$$

线性回归的目标是通过优化参数 $\mathbf{w}, b$，使模型预测值：

$$
\hat{y}_i = \mathbf{w}^T x_i + b
$$

尽可能逼近真实值 $y_i$。

通常通过最小化均方误差（MSE）来估计参数：

$$
\min_{\mathbf{w}, b} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$

线性回归是最基础的监督学习模型之一，也是神经网络的基本形式。
