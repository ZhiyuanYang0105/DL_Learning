向量
## 简单操作

$c = a + b \quad \text{where } c_i = a_i + b_i$

$c = \alpha \cdot b \quad \text{where } c_i = \alpha b_i$

$c = \sin a \quad \text{where } c_i = \sin a_i$

---

## 长度

$$
\|a\|_2 =
\left(
\sum_{i=1}^{m} a_i^2
\right)^{\frac{1}{2}}
$$

$$
\|a\| \ge 0 \quad \text{for all } a
$$

$$
\|a + b\| \le \|a\| + \|b\|
$$

$$
\|\alpha b\| = |\alpha| \cdot \|b\|
$$
​	
 ## 点乘（内积）

向量 $a$ 和 $b$ 的点乘定义为：

$$
a^T b = \sum_{i=1}^{n} a_i b_i
$$

---

## 正交

如果

$$
a^T b = \sum_{i=1}^{n} a_i b_i = 0
$$

则称向量 $a$ 与 $b$ 正交（orthogonal）。

# 矩阵范数（Matrix Norm）

---

## 1️⃣ 基本关系

设

$$
c = A b
$$

则有

$$
\|c\| \le \|A\| \, \|b\|
$$

这表示：  
矩阵 $A$ 作为一个线性变换，对向量 $b$ 的长度放大是有上界的。

---

## 2️⃣ 矩阵范数的定义

矩阵范数定义为：

> 满足
>
> $$
> \|A b\| \le \|A\| \|b\|
> $$
>
> 的**最小常数**。

等价地，可以写成：

$$
\|A\| = \sup_{b \ne 0}
\frac{\|A b\|}{\|b\|}
$$

含义：

矩阵范数表示线性变换的**最大伸缩倍数**。

---

## 3️⃣ 常见矩阵范数

### ① 诱导 2-范数（Operator Norm）

$$
\|A\|_2 =
\sup_{b \ne 0}
\frac{\|A b\|_2}{\|b\|_2}
$$

它等于矩阵的**最大奇异值**。

---

### ② Frobenius 范数

$$
\|A\|_F =
\left(
\sum_{i,j} A_{ij}^2
\right)^{\frac{1}{2}}
$$

含义：

把矩阵看作一个大向量，计算其欧几里得长度。

---

## 4️⃣ 直观理解

- 向量范数：向量有多长  
- 矩阵范数：矩阵最多能把向量拉长多少倍  
- 若 $\|A\| > 1$：可能放大  
- 若 $\|A\| < 1$：整体收缩  

---

## 5️⃣ 与深度学习的关系

神经网络一层：

$$
x_{l+1} = W x_l
$$

若

$$
\|W\| > 1
$$

可能导致梯度爆炸；

若

$$
\|W\| < 1
$$

可能导致梯度消失。





# 特殊矩阵

---

## 1️⃣ 对称矩阵（Symmetric Matrix）

矩阵 $A$ 称为对称矩阵，如果：

$$
A = A^T
$$

等价地：

$$
A_{ij} = A_{ji}
$$

含义：

矩阵关于主对角线对称。

---

## 2️⃣ 反对称矩阵（Skew-Symmetric Matrix）

矩阵 $A$ 称为反对称矩阵，如果：

$$
A = -A^T
$$

等价地：

$$
A_{ij} = -A_{ji}
$$

性质：

- 主对角线元素必须为 0
- $A_{ii} = 0$

---

## 3️⃣ 正定矩阵（Positive Definite Matrix）

向量的平方范数：

$$
\|x\|^2 = x^T x \ge 0
$$

推广到矩阵：

$$
x^T A x \ge 0
$$

若对所有 $x \ne 0$ 都满足：

$$
x^T A x > 0
$$

则称 $A$ 为 **正定矩阵**。

---

## 4️⃣ 直观理解

- 对称矩阵：关于主对角线镜像对称  
- 反对称矩阵：对角线为 0，上下符号相反  
- 正定矩阵：二次型始终非负（或严格正）

---

## 5️⃣ 重要性质

对于对称矩阵：

- 所有特征值为实数  
- 若所有特征值 > 0，则矩阵正定  
- 若所有特征值 ≥ 0，则矩阵半正定  

---

## 6️⃣ 与深度学习的关系

- 协方差矩阵一定是半正定  
- Hessian 矩阵正定 ⇒ 极小值点  
- 正定矩阵保证优化问题是凸的

# 正交矩阵与置换矩阵

---

## 1️⃣ 正交矩阵（Orthogonal Matrix）

矩阵 $U$ 称为正交矩阵，如果：

- 所有行向量彼此正交  
- 所有行向量长度为 1  

数学表达为：

$$
\sum_j U_{ij} U_{kj} = \delta_{ik}
$$

其中 $\delta_{ik}$ 是 Kronecker delta：

$$
\delta_{ik} =
\begin{cases}
1, & i = k \\
0, & i \ne k
\end{cases}
$$

---

### 等价表达

正交矩阵可以写成：

$$
U U^T = I
$$

等价地：

$$
U^T U = I
$$

因此：

$$
U^{-1} = U^T
$$

---

### 直观理解

正交矩阵表示：

- 旋转
- 翻转

不会改变向量长度。

$$
\|Ux\| = \|x\|
$$

---

## 2️⃣ 置换矩阵（Permutation Matrix）

置换矩阵 $P$ 定义为：

$$
P_{ij} = 1 \quad \text{当且仅当 } j = \pi(i)
$$

其中 $\pi$ 是一个排列（permutation）。

---

### 性质

- 每一行只有一个 1  
- 每一列只有一个 1  
- 其余元素为 0  

---

### 重要性质

置换矩阵是正交矩阵：

$$
P P^T = I
$$

因此：

$$
P^{-1} = P^T

# 二、特征值与特征向量（Eigenvalues & Eigenvectors）

## 定义

对于矩阵 $A$，如果存在非零向量 $x$ 和标量 $\lambda$ 使得：

$$
A x = \lambda x
$$

则：

- $\lambda$ 称为特征值  
- $x$ 称为特征向量  

---

## 含义

矩阵作用在特征向量上：

- 不改变方向  
- 只改变长度  

放大或缩小倍数为：

$$
\lambda
$$

---

## 求特征值

由

$$
A x = \lambda x
$$

得到

$$
(A - \lambda I)x = 0
$$

要有非零解，必须满足：

$$
\det(A - \lambda I) = 0
$$

这称为特征方程。

---

## 几何意义

- 特征向量是“不会被旋转的方向”
- 特征值表示沿该方向的伸缩比例

---

## 与矩阵范数的关系

矩阵的 2-范数等于：

$$
\|A\|_2 = \text{最大奇异值}
$$

若 $A$ 是对称矩阵，则：

$$
\|A\|_2 = \max |\lambda_i|
$$

---

## 正定矩阵与特征值

对于对称矩阵：

- 若所有特征值 > 0 → 正定  
- 若所有特征值 ≥ 0 → 半正定
