# 标量导数

---

## 一、基本函数的导数

$$
\begin{aligned}
\frac{d}{dx} a &= 0 \qquad (a \text{ 不是 } x \text{ 的函数}) \\[6pt]
\frac{d}{dx} x^n &= n x^{n-1} \\[6pt]
\frac{d}{dx} \exp(x) &= \exp(x) \\[6pt]
\frac{d}{dx} \log(x) &= \frac{1}{x} \\[6pt]
\frac{d}{dx} \sin(x) &= \cos(x)
\end{aligned}
$$

---

## 二、运算法则

### 1. 加法法则

$$
\frac{d}{dx}(u+v)
=
\frac{du}{dx}
+
\frac{dv}{dx}
$$

---

### 2. 乘法法则

$$
\frac{d}{dx}(uv)
=
\frac{du}{dx}v
+
\frac{dv}{dx}u
$$

---

### 3. 链式法则

设
$$
y=f(u), \quad u=g(x)
$$

则

$$
\frac{dy}{dx}
=
\frac{dy}{du}
\frac{du}{dx}
$$

# 亚导数（Subgradient）

---

## 一、将导数推广到不可微函数

考虑函数

$$
y = |x|
$$

在 $x \neq 0$ 处可导：

$$
\frac{\partial |x|}{\partial x}
=
\begin{cases}
1, & x > 0 \\[6pt]
-1, & x < 0
\end{cases}
$$

但在 $x = 0$ 处不可导。

---

## 二、亚导数的定义

在不可导点 $x = 0$ 处，

$$
\frac{\partial |x|}{\partial x}
=
\begin{cases}
1, & x > 0 \\[6pt]
-1, & x < 0 \\[6pt]
a, & x = 0, \quad a \in [-1,1]
\end{cases}
$$

即在 $x=0$ 处，导数可以取区间 $[-1,1]$ 内的任意值。

我们称：

$$
\partial |x| \big|_{x=0} = [-1,1]
$$

这就是 **亚梯度集合（subdifferential）**。

---

## 三、另一个例子：ReLU 函数

定义

$$
\mathrm{ReLU}(x) = \max(x,0)
$$

其导数（亚导数）为：

$$
\frac{\partial}{\partial x} \max(x,0)
=
\begin{cases}
1, & x > 0 \\[6pt]
0, & x < 0 \\[6pt]
a, & x = 0, \quad a \in [0,1]
\end{cases}
$$

---

## 四、核心思想

对于凸函数：

- 在可导点，亚导数就是普通导数
- 在不可导点，亚导数是一个区间（集合）

亚导数本质上是：

> 所有能够作为该点“支撑超平面”斜率的集合

# 梯度与向量/矩阵求导（Matrix Calculus）

---

# 一、从标量到向量

我们将导数从标量推广到向量。

设：

$$
x =
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
\in \mathbb{R}^n
$$

---

## 1️⃣ 标量函数对向量求导

若

$$
y = f(x)
$$

其中 $y \in \mathbb{R}$，则

$$
\frac{\partial y}{\partial x}
=
\begin{bmatrix}
\frac{\partial y}{\partial x_1},
\frac{\partial y}{\partial x_2},
\cdots,
\frac{\partial y}{\partial x_n}
\end{bmatrix}
\in \mathbb{R}^{1 \times n}
$$

这称为：

> 行向量形式（分子布局 notation）

梯度（列向量形式）为：

$$
\nabla_x y
=
\begin{bmatrix}
\frac{\partial y}{\partial x_1} \\
\frac{\partial y}{\partial x_2} \\
\vdots \\
\frac{\partial y}{\partial x_n}
\end{bmatrix}
\in \mathbb{R}^{n}
$$

---

## 2️⃣ 示例

### （1）二次函数

$$
f(x_1,x_2) = x_1^2 + 2x_2^2
$$

则

$$
\frac{\partial f}{\partial x}
=
\begin{bmatrix}
2x_1 & 4x_2
\end{bmatrix}
$$

梯度方向为：

$$
\nabla f =
\begin{bmatrix}
2x_1 \\
4x_2
\end{bmatrix}
$$

梯度与等高线正交。

---

# 二、向量函数对向量求导（Jacobian）

若

$$
y =
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_m
\end{bmatrix}
\in \mathbb{R}^m
$$

则

$$
\frac{\partial y}{\partial x}
=
\begin{bmatrix}
\frac{\partial y_1}{\partial x_1} & \cdots & \frac{\partial y_1}{\partial x_n} \\
\frac{\partial y_2}{\partial x_1} & \cdots & \frac{\partial y_2}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial y_m}{\partial x_1} & \cdots & \frac{\partial y_m}{\partial x_n}
\end{bmatrix}
\in \mathbb{R}^{m \times n}
$$

这称为：

> Jacobian 矩阵

---

# 三、基本求导公式（向量情形）

设：

- $x \in \mathbb{R}^n$
- $a \in \mathbb{R}$
- $\mathbf{a} \in \mathbb{R}^m$
- $A \in \mathbb{R}^{m \times n}$

且它们均不依赖于 $x$。

---

## 1️⃣ 常数

$$
y = a
\quad \Rightarrow \quad
\frac{\partial y}{\partial x} = 0
$$

---

## 2️⃣ 恒等映射

$$
y = x
\quad \Rightarrow \quad
\frac{\partial y}{\partial x} = I
$$

---

## 3️⃣ 线性变换

$$
y = Ax
\quad \Rightarrow \quad
\frac{\partial y}{\partial x} = A
$$

---

## 4️⃣ 转置形式

$$
y = x^T A
\quad \Rightarrow \quad
\frac{\partial y}{\partial x} = A^T
$$

---

# 四、运算法则

---

## 1️⃣ 标量乘法

$$
y = a u
\quad \Rightarrow \quad
\frac{\partial y}{\partial x}
=
a \frac{\partial u}{\partial x}
$$

---

## 2️⃣ 矩阵乘法

$$
y = Au
\quad \Rightarrow \quad
\frac{\partial y}{\partial x}
=
A \frac{\partial u}{\partial x}
$$

---

## 3️⃣ 加法法则

$$
y = u + v
$$

$$
\frac{\partial y}{\partial x}
=
\frac{\partial u}{\partial x}
+
\frac{\partial v}{\partial x}
$$

---

## 4️⃣ 内积

$$
y = \langle u, v \rangle = u^T v
$$

$$
\frac{\partial y}{\partial x}
=
u^T \frac{\partial v}{\partial x}
+
v^T \frac{\partial u}{\partial x}
$$

---

# 五、推广到矩阵

若

$$
X \in \mathbb{R}^{n \times k}
$$

---

## 1️⃣ 标量对矩阵求导

$$
\frac{\partial y}{\partial X}
\in \mathbb{R}^{k \times n}
$$

---

## 2️⃣ 向量对矩阵求导

$$
\frac{\partial y}{\partial X}
\in \mathbb{R}^{m \times n}
$$

---

## 3️⃣ 矩阵对矩阵求导

$$
Y \in \mathbb{R}^{m \times l}
$$

则

$$
\frac{\partial Y}{\partial X}
\in \mathbb{R}^{m \times l \times n \times k}
$$

这是一个四阶张量。

---

# 六、维度总结

| 输入 | 输出 | 导数形状 |
|------|------|----------|
| 标量 → 标量 | | 标量 |
| 向量 → 标量 | | 行向量 $(1,n)$ |
| 向量 → 向量 | | 矩阵 $(m,n)$ |
| 矩阵 → 标量 | | 矩阵 |
| 矩阵 → 矩阵 | | 四阶张量 |

---

# 七、核心思想

1. 标量函数对向量求导得到梯度
2. 向量函数对向量求导得到 Jacobian
3. 矩阵函数对矩阵求导得到高阶张量
4. 梯度方向与等高线正交
5. 线性函数的导数仍然是线性映射

# 八、最重要的几个公式（机器学习必备）

$$
\nabla_x (x^T x) = 2x
$$

$$
\frac{\partial}{\partial x}(Ax) = A
$$

$$
\frac{\partial}{\partial x}(x^T A) = A^T
$$

$$
\frac{\partial}{\partial x}(u^T v)
=
u^T \frac{\partial v}{\partial x}
+
v^T \frac{\partial u}{\partial x}
$$
